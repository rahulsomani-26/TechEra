```python
from threading import Thread
import time 
def func():
    start = time.time()
    print('Sleeping for 2 secs')
    time.sleep(2)
    print('Sleeping for 3 secs')
    time.sleep(3)
    end = time.time()
    print('Time taken = {}'.format(end-start))

    # create threads
    
t1 = Thread(target=func)
t2 = Thread(target=func)
    
    # start threads
    
t1.start()
t2.start()
    
    
    # Join the the thread
    
t1.join()
t2.join()
    
print(" End of the program".center(100,'*'))    
    
    
    
```

```python
from threading import Thread
import time 
def func():
    start = time.time()
    print('Sleeping for 2 secs')
    time.sleep(2)
    print('Sleeping for 3 secs')
    time.sleep(3)
    end = time.time()
    print('Time taken = {}'.format(end-start))

    # create threads
    
    t1 = Thread(target=func)
    t2 = Thread(target=func)

        # start threads

    t1.start()
    t2.start()


        # Join the the thread

    t1.join()
    t2.join()

    print(" End of the program".center(100,'*'))    
    
    
    
```

```python
import time
from threading import Thread
start = time.time()

def my_func():
    print("Inside {}".format(__name__))
    time.sleep(1)
    print("Calculating")
    time.sleep(1)
    
t1 = Thread(target=my_func)
t2 = Thread(target=my_func)
t3 = Thread(target=my_func)
t4 = Thread(target=my_func)

t1.start()
t2.start()
t3.start()
t4.start()

t1.join()
t2.join()
t3.join()
t4.join()

end = time.time()
print('Total time taken by = {}'.format(end-start))
    
```
